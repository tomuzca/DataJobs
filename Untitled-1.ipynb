```python
import pandas as pd


ruta_archivo_csv = 'jobs_in_data.csv' 


df = pd.read_csv(ruta_archivo_csv)


df.head()
```

```python
df.columns
```

```python
df.info()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


promedio_salario_por_ubicacion = df.groupby('company_location')['salary_in_usd'].mean()


plt.figure(figsize=(18, 6))
promedio_salario_por_ubicacion.plot(kind='bar', color='skyblue')
plt.title('Promedio de Salary en USD por Ubicacin de la Empresa')
plt.xlabel('Ubicacin de la Empresa')
plt.ylabel('Promedio de Salary en USD')
plt.xticks(rotation=45, ha='right') 
plt.tight_layout()


plt.show()
```

```python
promedio_salario_por_ubicacion = df.groupby('job_title')['salary_in_usd'].mean()


plt.figure(figsize=(18, 6))
promedio_salario_por_ubicacion.plot(kind='bar', color='skyblue')
plt.title('Promedio de Salary en USD por Ubicacin de la Empresa')
plt.xlabel('Job title')
plt.ylabel('Promedio de Salary en USD')
plt.xticks(rotation=45, ha='right')  
plt.tight_layout()


plt.show()
```

```python
import pandas as pd


promedio_salario_por_job_title = df.groupby('job_title')['salary_in_usd'].mean()


df_promedio_salario = pd.DataFrame({'job_title': promedio_salario_por_job_title.index, 'promedio_salary_in_usd': promedio_salario_por_job_title.values})


df_promedio_salario_ordenado = df_promedio_salario.sort_values(by='promedio_salary_in_usd', ascending=False)


df_promedio_salario_ordenado.tail(10)
```

```python
import pandas as pd
import matplotlib.pyplot as plt


sum_employee_residence = df.groupby('employee_residence').size().reset_index(name='counts')

# Graficar la suma de cada 'employee_residence'
plt.figure(figsize=(10, 6))
plt.bar(sum_employee_residence['employee_residence'], sum_employee_residence['counts'], color='skyblue')
plt.xlabel('Employee Residence')
plt.ylabel('Suma')
plt.title('Suma de cada Employee Residence')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
sum_employee_residence = df.groupby('employee_residence').size().reset_index(name='counts')


sum_employee_residence_sorted = sum_employee_residence.sort_values(by='counts', ascending=False)


print("Tabla de los employee_residence que ms se repiten:")
print(sum_employee_residence_sorted.head(20))
```
