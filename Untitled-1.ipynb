```python
import pandas as pd

# Especifica la ruta del archivo CSV
ruta_archivo_csv = 'jobs_in_data.csv'  # Reemplaza 'ruta_del_archivo.csv' con la ruta real de tu archivo CSV

# Carga el archivo CSV en un DataFrame
df = pd.read_csv(ruta_archivo_csv)

# Muestra las primeras filas del DataFrame
print(df.head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Supongamos que ya tienes un DataFrame llamado 'dataframe' con las columnas 'salary_in_usd' y 'company_location'

# Agrupa por la ubicacin de la empresa y calcula el promedio del salario para cada ubicacin
promedio_salario_por_ubicacion = df.groupby('company_location')['salary_in_usd'].mean()

# Crea un grfico de barras para visualizar el promedio de salario por ubicacin
plt.figure(figsize=(12, 6))
promedio_salario_por_ubicacion.plot(kind='bar', color='skyblue')
plt.title('Promedio de Salary en USD por Ubicacin de la Empresa')
plt.xlabel('Ubicacin de la Empresa')
plt.ylabel('Promedio de Salary en USD')
plt.xticks(rotation=45, ha='right')  # Rotar las etiquetas del eje x para mayor legibilidad
plt.tight_layout()

# Muestra el grfico
plt.show()
```

```python
promedio_salario_por_ubicacion = df.groupby('job_title')['salary_in_usd'].mean()

# Crea un grfico de barras para visualizar el promedio de salario por ubicacin
plt.figure(figsize=(12, 6))
promedio_salario_por_ubicacion.plot(kind='bar', color='skyblue')
plt.title('Promedio de Salary en USD por Ubicacin de la Empresa')
plt.xlabel('Job title')
plt.ylabel('Promedio de Salary en USD')
plt.xticks(rotation=45, ha='right')  # Rotar las etiquetas del eje x para mayor legibilidad
plt.tight_layout()

# Muestra el grfico
plt.show()
```
